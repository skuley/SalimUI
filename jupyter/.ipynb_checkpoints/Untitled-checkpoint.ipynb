```python
def check_cert_information(self, cert_lst):
        total_result = []
        for idx, ocr_lst in enumerate(cert_lst):
            ocr_num, ocr_word = self.check_num_str(ocr_lst)
            top_score = 0
            infor_result = {}
            gubun_dic = {}
            cert_name = self.cert_num_db['눈으로보고고치기']
            cert_num = self.cert_num_db['빼기기호없애기']
            for name_db, num_db in zip(cert_name, cert_num):
                point = 0
                score = 0
                if len(ocr_num) == len(num_db):
                    for i in range(len(ocr_num)):
                        if ocr_num[i] == num_db[i]:
                            point += 1
                    score = point / len(ocr_num)
                elif len(ocr_num) > len(num_db):
                    for i in range(len(num_db)):
                        if ocr_num[i] == num_db[i]:
                            point += 1
                    score = point / len(ocr_num)
                elif len(ocr_num) < len(num_db):
                    for i in range(len(ocr_num)):
                        if ocr_num[i] == num_db[i]:
                            point += 1
                    score = point / len(num_db)
                gubun_dic[num_db] = [score, name_db]
                if top_score == 100:
                    same_text = num_db
                    break
                if top_score < score:
                    top_score = score

            if len([k for k, v in gubun_dic.items() if v[0] == top_score]) != 1:
                infor_result['result'] = '인식실패'
            else:
                if top_score > self.standard / 100:
                    top_num = [k for k, v in gubun_dic.items() if v[0] == top_score][0]
                    top_text = [v[1] for k, v in gubun_dic.items() if v[0] == top_score][0]
                    infor_result['ocr_num'] = ocr_num
                    infor_result['db_num'] = top_num
                    infor_result['num_score'] = top_score
                    name_result = self.similar([ocr_word], top_text)
                    infor_result['ocr_name'] = ocr_word
                    infor_result['db_name'] = top_text
                    infor_result['name_score'] = name_result['score']
                    infor_result['incorrect'] = name_result['incorrect_letter']
                    infor_result['result'] = '승인'
                    infor_result['reason'] = ''
                else:
                    infor_result['reason'] = '인식결과가 기준보다 낮아요'
                    infor_result['result'] = '매칭실패'
            total_result.append(infor_result)
        return total_result

    def check_num_str(self, input_str):
        word = ''
        num = ''
        for string in input_str:
            if string.isdigit():
                num += string
            else:
                word += string
                word = word.strip()
        return (num, word)
```
