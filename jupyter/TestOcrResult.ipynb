```python
import cv2
import os
from PIL import Image
from threading import Thread
import time

ocr_image = []

def tracking():
    global ocr_image
    video_path = 'A:/salim/sample_videos/20211104_0941.mp4'
    video = cv2.VideoCapture(video_path)
    fps = video.get(cv2.CAP_PROP_FPS)
    while video.isOpened():
        ret, frame = video.read()
        print(len(ocr_image))
        if len(ocr_image) < 1:
            ocr_image.append(frame)
        key = cv2.waitKey(0)

        time.sleep(1)
    video.release()

def ocr():
    global ocr_image
    while True:
        if ocr_image:
            ocr_image.pop(0)
            time.sleep(2)

def main():
    tracking_thread = Thread(target=tracking, args=())
    tracking_thread.daemon = True
    tracking_thread.start()


    ocr_thread = Thread(target=ocr, args=())
    ocr_thread.daemon = True
    ocr_thread.start()


if __name__ == "__main__":
    if ocr_image:
        print(ocr_image[0].shape)
    main()

# Image.fromarray(frame)
```
